```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq, fftshift
from scipy.signal.windows import boxcar

# Gerar sinal janela retangular unitria
sinal = boxcar(50)

# Calcular a transformada de Fourier do sinal
transformada = fft(sinal, 2048) / (len(sinal) / 2.0)
transformada = np.abs(fftshift(transformada / abs(transformada).max()))
frequencias = np.linspace(-0.5, 0.5, len(transformada))

# Plotar o sinal no tempo
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(sinal)
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.title('Sinal Janela Unitria no Tempo com a funo BoxCar')
plt.ylim(0, 2)

# Plotar a transformada de Fourier
plt.subplot(2, 1, 2)
plt.plot(frequencias, transformada)
plt.xlabel('Frequncia (Hz)')
plt.ylabel('Amplitude')
plt.title('Transformada de Fourier do Sinal')

# Exibir os grficos
plt.tight_layout()
plt.show()
```
