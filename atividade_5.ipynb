##Detector de bordas horizontais

```python
import cv2
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
imagem = np.zeros((256, 256), dtype=np.uint8)
x1, y1 = 64, 64
x2, y2 = 192, 192
imagem[y1:y2, x1:x2] = 255
filtro = np.array([[1, 1, 1],
                   [0, 0, 0],
                   [-1,-1,-1]])
imagem_filtrada2 = signal.convolve2d(imagem, filtro, 'valid')
plt.subplot(1, 2, 1)
plt.imshow(imagem, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(imagem_filtrada2, cmap='gray')
plt.title('Imagem Filtrada')
plt.axis('off')
plt.tight_layout()
plt.show()
```

##Detector de bordas e de direção das bordas

```python
import cv2
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
imagem = np.zeros((256, 256), dtype=np.uint8)
x1, y1 = 64, 64
x2, y2 = 192, 192
imagem[y1:y2, x1:x2] = 255
filtro = np.array([[ -3-3j, 0-10j,  +3 -3j],
                   [-10+0j, 0+ 0j, +10 +0j],
                   [ -3+3j, 0+10j,  +3 +3j]])
imagem_filtrada3 = signal.convolve2d(imagem, filtro, 'valid')
plt.figure(figsize=(6.4, 7))
plt.subplot(2, 2, 1)
plt.imshow(imagem, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')
plt.subplot(2, 2, 2)
plt.imshow(np.absolute(imagem_filtrada3), cmap='gray')
plt.title('Bordas')
plt.axis('off')
plt.subplot(2, 2, 4)
plt.imshow(np.angle(imagem_filtrada3), cmap='hsv')
plt.title('Direção das Bordas')
plt.axis('off')
plt.tight_layout()
plt.show()
```

##Filtro de Sobel

```python
import cv2
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
imagem = np.zeros((256, 256), dtype=np.uint8)
x1, y1 = 64, 64
x2, y2 = 192, 192
imagem[y1:y2, x1:x2] = 255
sobel_x = cv2.Sobel(imagem, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(imagem, cv2.CV_64F, 0, 1, ksize=3)
sobel_xy = cv2.Sobel(imagem, cv2.CV_64F, 1, 1, ksize=3)
sobel_x = cv2.convertScaleAbs(sobel_x)
sobel_y = cv2.convertScaleAbs(sobel_y)
sobel_xy = cv2.convertScaleAbs(sobel_xy)
plt.figure(figsize=(6.4, 7))
plt.subplot(2, 2, 1)
plt.imshow(imagem, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')
plt.subplot(2, 2, 2)
plt.imshow(sobel_x, cmap='gray')
plt.title('Bordas Verticais (Sobel)')
plt.axis('off')
plt.subplot(2, 2, 3)
plt.imshow(sobel_xy, cmap='gray')
plt.title('Derivando X e Y')
plt.axis('off')
plt.subplot(2, 2, 4)
plt.imshow(sobel_y, cmap='gray')
plt.title('Bordas Horizontais (Sobel)')
plt.axis('off')
plt.tight_layout()
plt.show()
```

##Filtro Laplaciano

##Detector de quinas de Harris

```python
import cv2
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
imagem = np.zeros((256, 256), dtype=np.uint8)
x1, y1 = 64, 64
x2, y2 = 192, 192
imagem[y1:y2, x1:x2] = 255
imagem_harris = np.float32(imagem)
dst = cv2.cornerHarris(imagem_harris, blockSize=3, ksize=3, k=0.04)
dst = cv2.dilate(dst, None)
imagem_harris[dst > 0.01 * dst.max()] = 255
plt.figure(figsize=(6.4, 7))
plt.subplot(2, 2, 1)
plt.imshow(imagem, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')
plt.subplot(2, 2, 2)
plt.imshow(imagem_harris, cmap='gray')
plt.title('Imagem com Pontos de Interesse Harris')
plt.axis('off')
plt.subplot(2, 2, 3)
plt.imshow(dst, cmap='gray')
plt.title('Pontos de Interesse Harris')
plt.axis('off')
plt.tight_layout()
plt.show()
```

##Obtenção de pontos de interesse com o algoritmo SIFT

```python
import cv2
import numpy as np
import urllib.request
import matplotlib.pyplot as plt
image_url = 'https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png'
urllib.request.urlretrieve(image_url, 'lena_gray.png')
imagem = cv2.imread('lena_gray.png', cv2.IMREAD_GRAYSCALE)
altura, largura = imagem.shape[:2]
centro = (largura // 2, altura // 2)
angulo = 15
rotacao_matriz = cv2.getRotationMatrix2D(centro, angulo, 1)
imagem_rotacionada = cv2.warpAffine(imagem, rotacao_matriz, (largura, altura))
sift = cv2.SIFT_create()
keypoints = sift.detect(imagem, None)
keypoints_rotacao = sift.detect(imagem_rotacionada, None)
imagem_sift = cv2.drawKeypoints(imagem, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
imagem_sift_rotacao = cv2.drawKeypoints(imagem_rotacionada, keypoints_rotacao, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.figure()
plt.subplot(1,2,1)
plt.imshow(imagem_sift, cmap='gray')
plt.axis('off')
plt.title('Imagem Original')
plt.subplot(1,2,2)
plt.imshow(imagem_sift_rotacao, cmap='gray')
plt.axis('off')
plt.title('Imagem Rotacionada')
plt.tight_layout()
plt.show()
```

##Usando SIFT para identificar a rotação em imagem

```python
import cv2
import numpy as np
import urllib.request
import matplotlib.pyplot as plt
image_url = 'https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png'
urllib.request.urlretrieve(image_url, 'lena_gray.png')
imagem = cv2.imread('lena_gray.png', cv2.IMREAD_GRAYSCALE)
sift = cv2.SIFT_create()
keypoints = sift.detect(imagem, None)
imagem_sift = cv2.drawKeypoints(imagem, keypoints, None)
altura, largura = imagem.shape[:2]
centro = (largura // 2, altura // 2)
angulo = 15
rotacao_matriz = cv2.getRotationMatrix2D(centro, angulo, 1)
imagem_rotacionada = cv2.warpAffine(imagem, rotacao_matriz, (largura, altura))
keypoints_rotacionados = sift.detect(imagem_rotacionada, None)
imagem_sift_rotacionada = cv2.drawKeypoints(imagem_rotacionada, keypoints_rotacionados, None)
pontos_originais = np.float32([keypoint.pt for keypoint in keypoints]).reshape(-1, 1, 2)
pontos_rotacionados = np.float32([keypoint.pt for keypoint in keypoints_rotacionados]).reshape(-1, 1, 2)
pontos_originais_transformados = cv2.transform(pontos_originais, rotacao_matriz)
for i in range(min(len(pontos_originais), len(pontos_rotacionados))):
    x1, y1 = pontos_originais[i][0]
    x2, y2 = pontos_rotacionados[i][0]
    x1_transformado, y1_transformado = pontos_originais_transformados[i][0]
    plt.plot([x1, x1_transformado], [y1, y1_transformado], 'r-', lw=1)
plt.axis('off')
plt.imshow(imagem_sift, cmap='gray')
plt.axis('off')
plt.imshow(imagem_sift_rotacionada, cmap='gray', alpha=0.5)
plt.axis('off')
plt.tight_layout()
plt.show()
```

##Identificando pontos de interesse com o algoritmo ORB

```python
import cv2
import numpy as np
import urllib.request
import matplotlib.pyplot as plt
image_url = 'https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png'
urllib.request.urlretrieve(image_url, 'lena_gray.png')
imagem = cv2.imread('lena_gray.png', cv2.IMREAD_GRAYSCALE)
imagem_filtrada = cv2.GaussianBlur(imagem,(7,7),cv2.BORDER_DEFAULT)
orb = cv2.ORB_create()
keypoints = orb.detect(imagem, None)
keypoints_filtro = orb.detect(imagem_filtrada, None)
imagem_orb = cv2.drawKeypoints(imagem, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
imagem_orb_filtro = cv2.drawKeypoints(imagem_filtrada, keypoints_filtro, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.figure()
plt.subplot(1,2,1)
plt.imshow(imagem_orb, cmap='gray')
plt.axis('off')
plt.title('Imagem Original')
plt.subplot(1,2,2)
plt.imshow(imagem_orb_filtro, cmap='gray')
plt.axis('off')
plt.title('Imagem Filtrada')
plt.tight_layout()
plt.show()
```

##Obtendo informação sobre a rotação com o ORB

```python
import cv2
import numpy as np
import urllib.request
import matplotlib.pyplot as plt
image_url = 'https://www.math.hkust.edu.hk/~masyleung/Teaching/CAS/MATLAB/image/images/cameraman.jpg'
urllib.request.urlretrieve(image_url, 'lena_gray.png')
imagem = cv2.imread('lena_gray.png', cv2.IMREAD_GRAYSCALE)
orb = cv2.ORB_create()
keypoints = orb.detect(imagem, None)
imagem_orb = cv2.drawKeypoints(imagem, keypoints, None)
altura, largura = imagem.shape[:2]
centro = (largura // 2, altura // 2)
angulo = 45
rotacao_matriz = cv2.getRotationMatrix2D(centro, angulo, 1)
imagem_rotacionada = cv2.warpAffine(imagem, rotacao_matriz, (largura, altura))
keypoints_rotacionados = orb.detect(imagem_rotacionada, None)
imagem_orb_rotacionada = cv2.drawKeypoints(imagem_rotacionada, keypoints_rotacionados, None)
pontos_originais = np.float32([keypoint.pt for keypoint in keypoints]).reshape(-1, 1, 2)
pontos_rotacionados = np.float32([keypoint.pt for keypoint in keypoints_rotacionados]).reshape(-1, 1, 2)
pontos_originais_transformados = cv2.transform(pontos_originais, rotacao_matriz)
imagem_final = np.concatenate((imagem_orb, imagem_orb_rotacionada), axis=1)
plt.figure()
plt.imshow(imagem_final)
offset = imagem_orb.shape[1]
for i in range(min(len(pontos_originais), len(pontos_rotacionados)) // 10 + 30):
    i += 30
    x1, y1 = pontos_originais[i][0]
    x2, y2 = pontos_rotacionados[i][0]
    x1_transformado, y1_transformado = pontos_originais_transformados[i][0]
    x1_transformado += offset
    plt.plot([x1, x1_transformado], [y1, y1_transformado], 'r-', lw=1)
plt.axis('off')
plt.show()
```
