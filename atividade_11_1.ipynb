##01 - Calculo da Diferen√ßa entre Imagens

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

cap = cv2.VideoCapture('movie_001.mp4')

# Pegue o primeiro frame
frame_1 = 40
cap.set(cv2.CAP_PROP_POS_FRAMES, frame_1)
ret, prev_frame = cap.read()
prev_gray = cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY)

# Pegue o segundo frame
frame_2 = 43
current_position = cap.get(cv2.CAP_PROP_POS_FRAMES)
cap.set(cv2.CAP_PROP_POS_FRAMES, frame_2)
ret, frame = cap.read()
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

# Calcula a diferen√ßa entre os Frames
diff = cv2.absdiff(prev_gray, gray)

# Aplique um limiar para pegar o objeto em movimento
_, thresholded = cv2.threshold(diff, 25, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(prev_frame, cv2.COLOR_BGR2RGB))
plt.title(f'Imagem da C√¢mera do frame {frame_1}')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
plt.title(f'Imagem da C√¢mera do frame {frame_2}')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(diff, cmap='gray')
plt.title(f'Diferen√ßa entre os Frames')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(thresholded, cmap='gray')
plt.title(f'Limiar de refor√ßo')
plt.axis('off')

plt.tight_layout()
plt.show()

cap.release()
```
