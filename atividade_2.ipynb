```python
import cv2
import matplotlib.pyplot as plt
imagem = cv2.imread('btob.jpg', cv2.IMREAD_GRAYSCALE)
plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(imagem, cmap='gray', vmin=0, vmax=255)
plt.title('Imagem Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(imagem, cmap='gray', vmin=40, vmax=150)
plt.title('Limitando a escala do imshow')
plt.xlabel('Nveis de Intensidade')
plt.axis('off')
plt.tight_layout()
plt.show()
```
