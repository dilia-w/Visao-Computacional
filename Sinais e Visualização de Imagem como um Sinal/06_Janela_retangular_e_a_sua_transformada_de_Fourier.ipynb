```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Parâmetros do sinal
tempo_total = .5  # Duração total do sinal em segundos
amostragem = 100  # Taxa de amostragem em Hz

# Gerar vetor de tempo
tempo = np.linspace(0, tempo_total, int(tempo_total * amostragem), endpoint=False)

# Gerar sinal janela retangular unitária
sinal = np.ones(len(tempo))
vetor_zeros = np.zeros(300)
sinal2 = np.concatenate((sinal, vetor_zeros))

# Calcular a transformada de Fourier do sinal
transformada = fft(sinal2)
frequencias = fftfreq(len(sinal2), 1 / amostragem)

# Plotar o sinal no tempo
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(tempo, sinal)
plt.ylim(0, 2)
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.title('Sinal Janela Unitária no Tempo')

# Plotar a transformada de Fourier
plt.subplot(2, 1, 2)
plt.plot(frequencias, np.abs(transformada))
plt.xlabel('Frequência (Hz)')
plt.ylabel('Amplitude')
plt.title('Transformada de Fourier do Sinal')

# Exibir os gráficos
plt.tight_layout()
plt.show()
```
