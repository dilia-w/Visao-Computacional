##01 - Canais de Cores RGB

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagem = cv2.imread('wally.jpg')

imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

imagem_r = np.copy(imagem)
imagem_r[:, :, 1] = 0
imagem_r[:, :, 2] = 0

imagem_g = np.copy(imagem)
imagem_g[:, :, 0] = 0
imagem_g[:, :, 2] = 0

imagem_b = np.copy(imagem)
imagem_b[:, :, 0] = 0
imagem_b[:, :, 1] = 0

plt.figure()

plt.subplot(2, 2, 1)
plt.imshow(imagem)
plt.title('Imagem Colorida')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(imagem_r, cmap='gray')
plt.title('Canal Vermelho')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(imagem_g, cmap='gray')
plt.title('Canal Verde')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(imagem_b, cmap='gray')
plt.title('Canal Azul')
plt.axis('off')

plt.tight_layout()
plt.show()
```

##02 - Cores do Espao de Cor HSV

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagem = np.zeros([10, 181, 3])

for i in range(181):
    imagem[:, i, 0] = i

imagem[:, :, 1] = 255
imagem[:, :, 2] = 255

imagem = imagem.astype(np.uint8)
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_HSV2RGB)

plt.figure()
plt.imshow(imagem_rgb)
plt.title('Faixa de cores para o HSV pelo OpenCV')
plt.xlim([0, 180])
plt.yticks([])
plt.tight_layout()
plt.show()
```

##03 - Canais de Cores HSV

```python
import numpy as np
import matplotlib.pyplot as plt

imagem = cv2.imread('wally.jpg')

imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
imagem_hsv = cv2.cvtColor(imagem, cv2.COLOR_RGB2HSV)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.imshow(imagem)
plt.title('Imagem Colorida')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(imagem_hsv[:, :, 0], cmap='gray')
plt.title('Canal Matiz = imagem_hsv[:,:,0]')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(imagem_hsv[:, :, 1], cmap='gray')
plt.title('Canal Saturao = imagem_hsv[:,:,1]')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(imagem_hsv[:, :, 2], cmap='gray')
plt.title('Canal Valor = imagem_hsv[:,:,2]')
plt.axis('off')

plt.tight_layout()
plt.subplots_adjust(wspace=-0.15)
plt.show()
```
