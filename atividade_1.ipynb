```python
import cv2
import matplotlib.pyplot as plt
imagem = cv2.imread('felix.jpg', cv2.IMREAD_GRAYSCALE)
plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(imagem, cmap='gray', vmin=0, vmax=255)
plt.title('Imagem Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.hist(imagem.ravel(), bins=256, range=(0, 255), color='gray')
plt.title('Histograma da Imagem Original')
plt.xlabel('Níveis de Intensidade')
plt.ylabel('Frequência')
plt.xlim(0, 255)
plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
imagem = cv2.imread('felix.jpg', cv2.IMREAD_GRAYSCALE)
plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(imagem, cmap='gray', vmin=0, vmax=255)
plt.title('Imagem Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(imagem, cmap='gray', vmin=40, vmax=150)
plt.title('Limitando a escala do imshow')
plt.xlabel('Níveis de Intensidade')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
imagem = cv2.imread('felix.jpg', cv2.IMREAD_GRAYSCALE)
imagem_equalizada = cv2.equalizeHist(imagem)
plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(imagem, cmap='gray', vmin=0, vmax=255)
plt.title('Imagem Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(imagem_equalizada, cmap='gray', vmin=0, vmax=255)
plt.title('Equalização do Histograma')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
imagem = cv2.imread('felix.jpg', cv2.IMREAD_GRAYSCALE)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
imagem_equalizada = clahe.apply(imagem)
plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(imagem, cmap='gray', vmin=0, vmax=255)
plt.title('Imagem Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(imagem_equalizada, cmap='gray', vmin=0, vmax=255)
plt.title('Equalização do Histograma')
plt.axis('off')
plt.tight_layout()
plt.show()
```
