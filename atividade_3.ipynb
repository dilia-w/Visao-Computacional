```python
import cv2
import matplotlib.pyplot as plt
imagem = cv2.imread('estatua.jpg', cv2.IMREAD_GRAYSCALE)
imagem_equalizada = cv2.equalizeHist(imagem)
plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(imagem, cmap='gray', vmin=0, vmax=255)
plt.title('Imagem Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(imagem_equalizada, cmap='gray', vmin=0, vmax=255)
plt.title('Equalização do Histograma')
plt.axis('off')
plt.tight_layout()
plt.show()
```
