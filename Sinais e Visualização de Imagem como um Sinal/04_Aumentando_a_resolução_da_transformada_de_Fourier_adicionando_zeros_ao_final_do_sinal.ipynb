```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Parâmetros do sinal
amplitude = 1.0  # Amplitude do sinal
frequencia = 10  # Frequência do sinal em Hz
tempo_total = .5  # Duração total do sinal em segundos
amostragem = 100  # Taxa de amostragem em Hz
vetor_zeros = np.zeros(1000)

# Gerar vetor de tempo
tempo = np.linspace(0, tempo_total, int(tempo_total * amostragem), endpoint=False)

# Gerar sinal senoidal
sinal = amplitude * np.sin(2 * np.pi * frequencia * tempo)
sinal2 = np.concatenate((sinal, vetor_zeros))

# Calcular a transformada de Fourier do sinal
transformada1 = fft(sinal)
frequencias1 = fftfreq(len(sinal), 1 / amostragem)
transformada2 = fft(sinal2)
frequencias2 = fftfreq(len(sinal2), 1 / amostragem)

# Plotar o sinal no tempo
plt.figure(figsize=(6.4, 7))

plt.subplot(3, 1, 1)
plt.plot(tempo, sinal)
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.title('Sinal no Tempo')

# Plotar a transformada de Fourier do sinal original
plt.subplot(3, 1, 2)
plt.plot(frequencias1, np.abs(transformada1))
plt.xlabel('Frequência (Hz)')
plt.ylabel('Amplitude')
plt.title('Transformada de Fourier do Sinal Original')

# Plotar a transformada de Fourier do sinal com zeros
plt.subplot(3, 1, 3)
plt.plot(frequencias2, np.abs(transformada2))
plt.xlabel('Frequência (Hz)')
plt.ylabel('Amplitude')
plt.title('Transformada de Fourier do Sinal com Zeros')

# Exibir os gráficos
plt.tight_layout()
plt.show()
```
